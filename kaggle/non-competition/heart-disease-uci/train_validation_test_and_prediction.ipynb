```python
def k_fold_cross_validation(model, dataset, k=10):  
    y = dataset['Disease']
    X = dataset.drop(['Disease'], axis=1)

    X_train, X_test, y_train, y_test = tts(X, y, test_size=0.1)

    kf = KFold(n_splits=k)
  
    accuracies = []
    X, y = X_train.values, y_train.values
  
    for train_index, test_index in kf.split(X):
        model.fit(X[train_index], y[train_index])
        accuracies.append(model.score(X[test_index], y[test_index]))
    
    print(accuracies)
  
    train_accuracy = model.score(X_train, y_train)
    test_accuracy = model.score(X_test, y_test)
  
    return mean(accuracies), train_accuracy, test_accuracy
  
```

```python
clf = KNeighborsClassifier(n_neighbors=3)

accuracy, train_accuracy, test_accuracy = k_fold_cross_validation(clf, dataset, k=10)
print(f"10-fold: {accuracy}")
print(f"train accuracy: {train_accuracy}")
print(f"test accuracy: {test_accuracy}")
```

```python
clf = DecisionTreeClassifier()

accuracy, train_accuracy, test_accuracy = k_fold_cross_validation(clf, dataset, k=10)
print(f"10-fold: {accuracy}")
print(f"train accuracy: {train_accuracy}")
print(f"test accuracy: {test_accuracy}")
```

```python
clf = RandomForestClassifier()

accuracy, train_accuracy, test_accuracy = k_fold_cross_validation(clf, dataset, k=10)
print(f"10-fold: {accuracy}")
print(f"train accuracy: {train_accuracy}")
print(f"test accuracy: {test_accuracy}")
```

```python
clf = MLPClassifier()

accuracy, train_accuracy, test_accuracy = k_fold_cross_validation(clf, dataset, k=10)
print(f"10-fold: {accuracy}")
print(f"train accuracy: {train_accuracy}")
print(f"test accuracy: {test_accuracy}")
```

## Scale

$ a(x_j) = \frac{x_i - min(x_j)}{max(x_i) - min(x_i)} $

```python
# scale the dataset
scaler = MinMaxScaler()
scaled_dataset = pd.DataFrame(scaler.fit_transform(dataset), columns=dataset.columns)
scaled_dataset.head()
```

```python
# reduce attributes
reduced_dataset = scaled_dataset.drop(['BloodPress', 'SerumChol', 'FastBloodSugar', 'ElectroCardioRest'], axis=1)
reduced_dataset.head()
```

```python
y = reduced_dataset["Disease"]
X = reduced_dataset.drop(["Disease"], axis=1)

# 10% destinado ao teste e os outros 90% para treino e validação usando k-fold cross validation
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.1)

X, y = X_train.values, y_train.values

n_fold = 10
```

```python
#knn

better_config = {}

#max_accuracy
max_acc = 0

for metric in ["euclidean", "manhattan", "chebyshev", "minkowski"]:
  
    for k in range(1, 15, 2):
    
        for fold in range(2, n_fold+1):
      
            kf = KFold(n_splits=fold)
      
            knn = KNeighborsClassifier(n_neighbors=k, metric=metric)
      
            results = []
            for train_index, valid_index in kf.split(X):
                knn.fit(X[train_index], y[train_index])
                results.append(knn.score(X[valid_index], y[valid_index]))
        
            if mean(results) > max_acc and np.std(results) < 0.1:
                max_acc = mean(results)
                better_config["metric"] = metric
                better_config["k"] = k
                better_config["acc"] = max_acc
        
            print(f"metric: {metric}, k: {k}, acc: {np.mean(results)}, std: {np.std(results)}")

print(better_config)
```

```python
better_config = {}

max_acc = 0

for criterion in ["gini", "entropy"]: # etropy to informatio gain
  
    for splitter in ["best", "random"]:
    
        for fold in range(2, n_fold+1):
      
            kf = KFold(n_splits=fold)
      
            dtree = DecisionTreeClassifier(criterion=criterion, splitter=splitter)
      
            results = []
            for train_index, valid_index in kf.split(X):
                dtree.fit(X[train_index], y[train_index])
                results.append(dtree.score(X[valid_index], y[valid_index]))
        
            if mean(results) > max_acc and np.std(results) < 0.1:
                max_acc = mean(results)
                better_config["criterion"] = criterion
                better_config["splitter"] = splitter
                better_config["acc"] = max_acc
        
            print(f"criterion: {criterion}, splitter: {splitter}, acc: {np.mean(results)}, std: {np.std(results)}")
            
print(better_config)
```

```python
better_config = {}

max_acc = 0

for n_estimators in range(10, 101):
  
    for criterion in ["gini", "entropy"]:
    
        for fold in range(2, n_fold+1):
      
            kf = KFold(n_splits=fold)
      
            rf = RandomForestClassifier(n_estimators=n_estimators, criterion=criterion)
      
            results = []
            for train_index, valid_index in kf.split(X):
                rf.fit(X[train_index], y[train_index])
                results.append(dtree.score(X[valid_index], y[valid_index]))
        
            if mean(results) > max_acc and np.std(results) < 0.1:
                max_acc = mean(results)
                better_config["criterion"] = criterion
                better_config["n_estimators"] = n_estimators
                better_config["acc"] = max_acc
        
            print(f"criterion: {criterion}, n_estimators: {n_estimators}, acc: {np.mean(results)}, std: {np.std(results)}")

print(better_config)      
```

```python
better_config = {}

max_acc = 0

for activation in ["identity", "logistic", "tanh", "relu"]:
  
    for solver in ["lbfgs", "sgd", "adam"]:
    
        for fold in range(2, n_fold+1):
      
            kf = KFold(n_splits=fold)
      
            mlp = MLPClassifier(activation=activation, solver=solver)
      
            results=[]
            for train_index, valid_index in kf.split(X):
                mlp.fit(X[train_index], y[train_index])
                results.append(mlp.score(X[valid_index], y[valid_index]))
        
            if mean(results) > max_acc and np.std(results) < 0.1:
                max_acc = mean(results)
                better_config["activation"] = activation
                better_config["solver"] = solver
                better_config["acc"] = max_acc
        
            print(f"activation: {activation}, solver: {solver}, acc: {np.mean(results)}, std: {np.std(results)}")
    
print(better_config) 
```

```python
classifiers = {"knn":[], "dtree":[], "rf":[], "mlp":[]}

for i in range(30):
  
    #knn
    knn = KNeighborsClassifier(n_neighbors=13, metric="manhattan")
    knn.fit(X_train, y_train)
    classifiers["knn"].append(knn.score(X_test, y_test))
  
    #decision tree
    dtree = DecisionTreeClassifier(criterion="entropy", splitter="random")
    dtree.fit(X_train, y_train)
    classifiers["dtree"].append(dtree.score(X_test, y_test))
  
    #random forest
    rf = RandomForestClassifier(n_estimators=10, criterion="gini")
    rf.fit(X_train, y_train)
    classifiers["rf"].append(rf.score(X_test, y_test))
  
    #mlp
    mlp = MLPClassifier(activation="relu", solver="adam")
    mlp.fit(X_train, y_train)
    classifiers["mlp"].append(mlp.score(X_test, y_test))

print("knn: %.2f, dtree: %.2f, rf: %.2f, mlp: %.2f" % 
      (mean(classifiers["knn"]), mean(classifiers["dtree"]), mean(classifiers["rf"]), mean(classifiers["mlp"])))
  
```
